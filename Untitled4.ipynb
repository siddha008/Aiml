# iqr method to find and remove outlires

```python
q1=fish1['Length'].quantile(0.25)
q3=fish1['Length'].quantile(0.75)
iqr=q3-q1
```

```python
q1,q3,iqr
```

```python
upper_bound=q3+(1.5*iqr)
lower_bound=q1-(1.5*iqr)
lower_bound,upper_bound
```

```python
sns.boxplot(fish1['Length'])
```

```python
#finding the outliers
fish1.loc[(fish1["Length"]>upper_bound) | (fish1["Length"]<lower_bound)]
```

```python
#triming the data
new_data = fish1.loc[(fish1["Length"]<=upper_bound)&(fish1["Length"]>=lower_bound)]
print("before triming:",len(fish1))
print("after triming:",len(new_data))
print("outliers:",len(fish1)-len(new_data))
```

```python
new_data=fish1.copy()
new_data.loc[new_data["Length"]>upper_bound, 'Length'] = upper_bound
new_data.loc[new_data["Length"]<lower_bound, 'Length'] = lower_bound
```
